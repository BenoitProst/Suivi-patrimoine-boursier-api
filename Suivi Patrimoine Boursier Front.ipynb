```python
import plotly.graph_objects as go
```

```python
import requests
import pandas as pd
```

```python
# Remplace par l'URL de ton API
url = "https://apisuivipatrimoineboursier.benoitprost.xyz/api/valeurmarche"

# Effectuer la requête GET
response = requests.get(url)

# Vérifier si la requête est réussie
if response.status_code == 200:
    data = response.json()  # Convertir la réponse en JSON
    df = pd.DataFrame(data)  # Charger dans un DataFrame
    print(df.head())  # Afficher les premières lignes
else:
    print(f"Erreur {response.status_code} : {response.text}")
```

```python
# Création du graphique avec Plotly
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df["Date"], y=df["Valeur marché"],
    mode='lines',
    name='Valeur Marché',
    line=dict(color='blue')
))

fig.add_trace(go.Scatter(
    x=df["Date"], y=df["Montant investi euro"],
    mode='lines',
    name='Montant Investi',
    line=dict(color='red')
))

fig.update_layout(
    title="Valorisation Patrimoine",
    xaxis_title="Date",
    yaxis_title="Valeur (€)",
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=True),
    template="presentation"
)

fig.show()
```
